```python
#Getting the imports needed, Time and Random
#Random import for the ‘choice()‘ method which is used to pick random words from our word lists.
#Time import for the ‘sleep()‘ method which is used for delays wherever necessary.
import time as tm
import random as rnd

#Displaying the welcome message
print("Welcome to the Hangman Game By Badim04")

#Creating the Word List which would be used in the game, Girl names and Boy names
Girl_names = ['amanda', 'brittany', 'amy', 'susan', 'gabrielle', 'zendaya','katy', 'camila','soo','aisha']
Boy_names = ['ted', 'barney', 'toby', 'steven', 'gabriel', 'son', 'harry', 'freddy', 'max', 'hakeem']

#Variables for storing the game statistics
play_game = True
Guesslist = []
Guesses = []
group = ""
continue_game = "Y"

#Creating a variable to store the keyboard input of the name the user wants
your_name = input("Enter Your Name Here: ")
print("Hello " + your_name.capitalize() +", Good Luck On Your Guesses!")

#To pause a little for the user, since we imported time as tm so we just use tm.sleep()
tm.sleep(1)

#Just displaying some game info
print("The Hangman Game is about guessing correctly the word picked by the program.")
tm.sleep(1)
print("You can guess only one letter at a time. Press the 'Enter Key' after each guess.")
tm.sleep(2)

#Another Display
print("Let's Start the Hangman Game Already")
tm.sleep(1)

print()

#Next, we choose a group from the 2 group of words we have, girl names and boy names
#We use a while loop for this
while True:
    
    while True:
#If the group uppercase letter is equal to 'G' then the word list we are using is Girl_names in which a word from there would be picked randomly. Then imported random as rnd
        if group.upper() == 'G':
            wordlist = rnd.choice(Girl_names)
            break
            
#Or if it's uppercase 'B' then the word list is Boys_name
        elif group.upper() == 'B':
            wordlist = rnd.choice(Boy_names)
            break
            
#Or else user input option again if the user types any other key
        else:
            group = input("Please select a group for the word list: 'G' for Girl_names Or 'B' for Boy_names Or 'X' to just exit the game: ")

#And if user input is 'X' then exit the game
        if group.upper() == 'X':
            print("Come Back Soon!")
            play_game = False
            break
            
            
#We show the length of the word chosen by random to be guessed denoted with '_'
#convert wordlist to a list stored in another variable
    if play_game:
        the_wordlist = list(wordlist)

#Now the player cannot have unlimited attempts at guessing the right letter so we limit that to 2 times the length of the word so 2 guesses for 1 letter average
        guess_attempt = (len(wordlist)*2)

#Then we need where the word will stay, a placeholder initiated with an empty list
        Guesslist = []

        print()

#Utility function to print User Guess List
        def printGuessedLetter():
            print("Your Secret word is: " + ''.join(Guesslist))

        print("The number of allowed guesses for this word is:", guess_attempt)

#Next we use a for loop to loop through the the_wordlist and add(append) a '_' for each element in the list
        for wl in the_wordlist: 
            Guesslist.append(' _ ')
        printGuessedLetter()#This is used for everytime we want to print our guessed letter

        
#A while loop to keep asking for letter inputs from the player
#It will first display a print to accept the user input of letter guessed
        while True: 
        
            print()
            
#Next, we ask the player to guess and enter a letter 
            print("Guess a letter that can be in the chosen word:")
            letter = input()#A variable called letter to store the user keyboard input
        
#So if letter is already been previously selected then it would display already guessed this letter
            if letter in Guesses:
                print("This letter has already been guessed by you.")

#Or else reduce the attempt by 1 and add the letter guessed to the right position in the word to be guessed
            else:
                guess_attempt -= 1
                Guesses.append(letter)
                if letter in the_wordlist:
                    print("You guessed right")
                    if guess_attempt > 1:#If attempts left are more than 1 then print the amount of attempts left
                        print("You have",guess_attempt, "guesses left")
                    if guess_attempt == 1:#If attempt left is equal to 1 then print 1 guess left
                        print("You have",guess_attempt, "guess left")
                        

#looping through the chosen word and inserting the guessed letter in the right index position
                    for index in range(len(the_wordlist)):
                        if letter == the_wordlist[index]:
                            letterIndex = index
                            Guesslist[letterIndex] = letter.upper()
                    printGuessedLetter()

#else wrong and show number attempts left
                else:
                    print("Nahhh, Wrong Guess.")
                    if guess_attempt > 1:
                        print("You have",guess_attempt, "guesses left")
                    if guess_attempt == 1:
                        print("You have",guess_attempt, "guess left")
                    printGuessedLetter()
                    
            print()
            
#We will join the ‘Guesslist’ variable and store it in the variable ‘alllist’.
            all_list = ''.join(Guesslist)
    
#If the user input and the guesslist match, the player will win
            if all_list.upper() == wordlist.upper():
                print("Yep, That's the correct word!")
                break
                
#Else if the guesses finish first then the player loses n it displays the correct word
            elif guess_attempt == 0:
                print("You have no more Guesses left and you lost the game!")
                print("Try Again!")
                print("The Chosen Word was: " + wordlist.upper())
                break
                
        print()    
        
#We prompt the player to check if they want to play again, if yes then we give it a group to choose from again
        continue_game = input("Do you want to play again? Y for Yes, I wanna play again, any other key to quit: ")
        if continue_game.upper() == 'Y':
            print()
            group = input("Please select a group: G for Girl names / B for Boy names: ")
            Guesslist = []
            Guesses = []
            play_game = True
            
#Else the player doesnt want to play again then Byee
        else:
            print("Thanks, Come back soon!")
            break
            
    else:
        break
```
